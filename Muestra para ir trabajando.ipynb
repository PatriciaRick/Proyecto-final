```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
df1 = pd.read_excel("./1PA.xlsx'")
df2 = pd.read_excel("./2PA.xlsx'")
df3 = pd.read_excel("./3PA.xlsx'")
df4 = pd.read_excel("./4PA.xlsx'")
df5 = pd.read_excel("./5PA.xlsx'")
df6 = pd.read_excel("./6PA.xlsx'")
df7 = pd.read_excel("./7PA.xlsx'")
df8 = pd.read_excel("./8PA.xlsx'")
df9 = pd.read_excel("./9PA.xlsx'")
```

```python
dfs=(df1,df2,df3,df4,df5,df6,df7,df8,df9)
register_number=[]
for df in dfs: 
    register_number.append(df['Unnamed: 1'][7:])
```

```python
def info(register): 
    url= 'https://cima.aemps.es:443/cima/dochtml/ft/'+register+'/FT_'+register+'.html'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    x = soup.select('section')
    return [i.text for i in x]
```

```python
register_number[0][1:10]
```

```python
information=[]
for i in register_number[0][1:1000].values:
    drug_info=info(i)
    drug_info=' '.join(drug_info)
    information.append(drug_info)
```

```python
df=pd.DataFrame(information, columns=['information'])
```

```python
df
```

```python
indicaciones=[]
for index in range(len(df['information'])):
    
    start_indicaciones = df['information'][index].find('Indicaciones terapéuticas') 
    end_indicaciones = df['information'][index].find('Posología y forma de administración', start_indicaciones)
    indicaciones.append(df['information'][index][start_indicaciones:end_indicaciones])

df['indicaciones'] =  indicaciones
```

```python
reacciones_adversas=[]
for index in range(len(df['information'])):
    
    start_reacciones_adversas = df['information'][index].find('Indicaciones terapéuticas') 
    end_reacciones_adversas = df['information'][index].find('Posología y forma de administración', start_reacciones_adversas)
    reacciones_adversas.append(df['information'][index][start_reacciones_adversas:end_reacciones_adversas])

df['reacciones adversas'] =  reacciones_adversas
```
