```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_drugs_by_letter(letter):
    url = 'https://www.vademecum.es/medicamentos-'+letter+'_1'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    names= soup.find_all(target="_top")
    for index in range (len(names)):
        names[index] = names[index].string
    return names  
```

```python
abc=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
names=[]

for l in abc: 
    names.append(get_drugs_by_letter(l))

names = [item for n in names for item in n]
df=pd.DataFrame(names,columns=['names_info'])
```
