```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_drugs_by_letter(letter):
    url = 'https://www.vademecum.es/medicamentos-'+letter+'_1'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    names= soup.find_all(target="_top")
    for index in range (len(names)):
        names[index] = names[index].string
        
    return names  
```

```python
abc=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

drugs=[get_drugs_by_letter(l) for l in abc]

drug = [item for element in drugs for item in element]
```

```python
column1_names=[' '.join(re.findall('[A-Z]+', str(drug[index]))[0:-1]) for index in range(len(drug))]
```

```python
def get_drugs_web(letter):
    url = 'https://www.vademecum.es/medicamentos-'+letter+'_1'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    href= soup.find_all(target="_top")
    for index in range (len(href)):
        href[index] = href[index].attrs['href']
     
    return href
```

```python
get_drugs_web('a')
```

```python
urls=[]

urls=[get_drugs_web(l) for l in abc]
url = [item for element in urls for item in element]
```

```python
def obtain_principioactivo(drug_url):
    url='https://www.vademecum.es'+drug_url
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    x = soup.select('td', id="TMP_new_ajax")
    return x
```

```python
principio_activo=[obtain_principioactivo('/medicamento-acarbosa++qualigen_38985') for element in url]
```
