```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_drugs_by_letter(letter):
    url = 'https://www.vademecum.es/medicamentos-'+letter+'_1'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    names= soup.find_all(target="_top")
    for index in range (len(names)):
        names[index] = names[index].string
        
    return names  
```

```python
abc=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

drugs=[get_drugs_by_letter(l) for l in abc]

drug = [item for element in drugs for item in element]
```

```python
def get_drugs_web(letter):
    url = 'https://www.vademecum.es/medicamentos-'+letter+'_1'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    href= soup.find_all(target="_top")
    for index in range (len(href)):
        href[index] = href[index].attrs['href']
     
    return href
```
