```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_drugs_by_letter(letter):
    url = 'https://www.clinicadam.com/medicamentos'+letter+'.html'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    
    names= soup.find_all('a')
    for index in range (len(names)):
        names[index] = names[index].string

    return names  
```

```python
abc=['','index-b','index-c','index-d','index-e','index-f','index-g','index-h','index-i','index-j','index-k','index-l',
     'index-m','index-n','index-o','index-p','index-q','index-r','index-s','index-t','index-u','index-v','index-w',
     'index-x','index-y','index-z']
```

```python
drugs=[]
for letter in abc:
    drugs.append(get_drugs_by_letter(letter)[31:113])
drugs = [item for element in drugs for item in element]
```

```python
df=pd.DataFrame(drugs, columns=['drugs'])
df.shape
```

```python
#eliminar todos los erróneos
errors=[(re.findall('[A-Z]', str(drugs[index]))) for index in range(len(drugs))]
errors=[item for element in errors for item in element]
errors.append(['Principios activos'])
               
#'Lesiones','Síntomas','Embarazo','Nutrición','Pruebas Diagnósticas','Procedimientos médicos y cirugías',
#'Intoxicaciones y envenenamientos','Temas Especiales de Salud','Temas de Salud','← Go to the home page',
#              'Condiciones de uso','Sitemap','None','Medicamentos','aquí','Condiciones de uso','Mapa del sitio',
#               'l','f','i','r'])
errors=[item for element in errors for item in element]
```

```python
df = df[~df['drugs'].isin(errors)]
```

```python
def drug_info(drug_url):
    url= 'https://www.clinicadam.com/medicamentos/'+drug_url+'.html'
    html= requests.get(url).content
    soup= BeautifulSoup(html, 'html')
    x = soup.select('p')
    return x
```

```python
info=[]
for drug in df['drugs']:
    info.append(drug_info(drug))
```
